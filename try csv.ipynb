```python
import os
import json
import pymysql
import csv

# 資料庫連線
connection = pymysql.connect(
    host = 'localhost',
    user = 'root',
    password = 'P@ssw0rd',
    database = 'moviedb',
    charset = 'utf8mb4',
    cursorclass = pymysql.cursors.DictCursor
)

cursor = connection.cursor()

with open('16.json', 'r', encoding="utf-8") as fp:
    strJSON = fp.read()
    listResult = json.loads(strJSON)
    fp.close()
    
def savemovie():
    a="movie_id,movie_title,belongs_to_collection,movie_popularity,movie_releasedate,movie_runtime,movie_budget,movie_revenue,movie_imdb_id\n"
    for itm in listResult:
        try:
            btc = itm['belongs_to_collection'] is not None
            a+='"{}","{}","{}","{}","{}","{}","{}","{}","{}"\n'.format(itm['id'],itm['original_title'],btc,itm['popularity'],itm['popularity'],itm['release_date'],itm['runtime'],itm['budget'],itm['revenue'],itm['imdb_id'])
        except Exception as e:
            pass
            continue
    fw=open('movie16.csv', 'w',encoding="utf-8")
    fw.write(a)
    fw.close()
    

def savegenres():
    b="movie_id,genre\n"
    for itm in listResult:
        try:
            list1=itm['genres']
            for itm1 in list1:
                b+='"{}","{}"\n'.format(itm['id'],itm1['name'])
        except Exception as e:
            pass
            continue
    fw=open('genres16.csv', 'w',encoding="utf-8")
    fw.write(b)
    fw.close()      

def savepc():
    c="movie_id,pc_id,movie_title,pc_name,pc_country\n"
    for itm in listResult:
        try:
            list1=itm['production_companies']
            for itm1 in list1:
                c+='"{}","{}","{}","{}","{}"\n'.format(itm['id'],itm1['id'],itm['original_title'],itm1['name'],itm1['origin_country'])
        except Exception as e:
            pass
            continue
    fw=open('pc16.csv', 'w',encoding="utf-8")
    fw.write(c)
    fw.close()        

def saveactor():
    d="movie_id,actor_id,actor_name,actor_gender,actor_character,popularity\n"
    for itm in listResult:
        try:
            list1=itm['credits']['cast']
            for itm1 in list1: 
                d+='"{}","{}","{}","{}","{}","{}"\n'.format(itm['id'],itm1['id'],itm1['original_name'],itm1['gender'],itm1['character'],itm1['popularity'])
        except Exception as e:
            pass
            continue
    fw=open('actor16.csv', 'w',encoding="utf-8")
    fw.write(d)
    fw.close()

def savecrew(): 
    f="movie_id,crew_id,crew_name,known_for_department,department,job\n"
    for itm in listResult:
        try:
            list1=itm['credits']['crew']
            for itm1 in list1: 
                f+='"{}","{}","{}","{}","{}","{}"\n'.format(itm['id'],itm1['id'],itm1['name'],itm1['known_for_department'],itm1['department'],itm1['job'])      
        except Exception as e:
            pass
            continue
    fw=open('crew16.csv', 'w',encoding="utf-8")
    fw.write(f)
    fw.close()

    
if __name__ == "__main__":
    savemovie()  
    savegenres()
    savepc()
    saveactor()
    savecrew()
```
